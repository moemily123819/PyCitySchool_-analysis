```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.columns
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
nbr_of_schools =len(school_data["school_name"].unique())
nbr_of_students = len(student_data["Student ID"])
total_budget =school_data["budget"].sum()
avg_math_score = student_data["math_score"].mean()
avg_rdg_score = student_data["reading_score"].mean()
passing_math=len(student_data.loc[student_data["math_score"] >=70]) /nbr_of_students *100
passing_rdg=len(student_data.loc[student_data["reading_score"] >=70])/nbr_of_students *100
overall_pass_rate = (avg_math_score + avg_rdg_score)/2

tot_budget = '{0:.2f}'.format(float(total_budget))

# Place all of the data found into a summary DataFrame
summary_table_df = pd.DataFrame({"Number of Schools": [nbr_of_schools],
                              "Number of Students": [nbr_of_students],
                              "Total Budget": ['$' + str(tot_budget)],
                              "Average Math Score": [avg_math_score],
                              "Average Reading Score": [avg_rdg_score],
                              "% Passing Math": [passing_math],
                              "% Passing Reading": [passing_rdg],
                              "Overall Passing Rate": [overall_pass_rate]
                                })
summary_table_df["Average Math Score"] = summary_table_df["Average Math Score"].map("{:.2f}".format)
summary_table_df["Average Reading Score"] = summary_table_df["Average Reading Score"].map("{:.2f}".format)
summary_table_df["% Passing Math"] = summary_table_df["% Passing Math"].map("{:.2f}%".format)
summary_table_df["% Passing Reading"] = summary_table_df["% Passing Reading"].map("{:.2f}%".format)
summary_table_df["Overall Passing Rate"] = summary_table_df["Overall Passing Rate"].map("{:.2f}%".format)

summary_table_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
summary_by_school_df = school_data_complete.groupby('school_name')

#school types
school_type = school_data.set_index('school_name')['type']

# total students by school
stu_per_school = school_data.set_index('school_name')['size']

# school budget
budget = school_data.set_index('school_name')['budget']

#per student budget
stu_budget = budget/stu_per_school

#number of students scored > 70 in math
math_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_school * 100 


#number of students scored > 70 in reading
reading_pass = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_school * 100 


summary_by_school_df = pd.DataFrame({
    "School Type" : school_type,
    "Total Students":stu_per_school,
    "Total School Budget":budget,
    "Per Student Budget":stu_budget,
    "Avg Reading Score": summary_by_school_df["reading_score"].mean(),
    "Avg Math Score": summary_by_school_df["math_score"].mean(),
     "% Passing Math" : math_pass,
     "% Passing Reading": reading_pass,
    "% Overall Pass Rate" : (math_pass + reading_pass) / 2,
    "% Overall Pass Rate w/out truncation" : (math_pass + reading_pass) / 2,
     "unformatted % Passing Math" : math_pass,
     "unformatted % Passing Reading": reading_pass,

})

summary_by_school_df["unformatted Avg Math Score"] = summary_by_school_df["Avg Math Score"]
summary_by_school_df["unformatted Avg Reading Score"] = summary_by_school_df["Avg Reading Score"]
summary_by_school_df["unformatted % Passing Math"] = summary_by_school_df["% Passing Math"]
summary_by_school_df["unformatted % Passing Reading"] = summary_by_school_df["% Passing Reading"]

summary_by_school_df["Avg Math Score"] = summary_by_school_df["Avg Math Score"].map("{:.2f}".format)
summary_by_school_df["Avg Reading Score"] = summary_by_school_df["Avg Reading Score"].map("{:.2f}".format)
summary_by_school_df["Total School Budget"] = summary_by_school_df["Total School Budget"].map("${:.2f}".format)
summary_by_school_df["Per Student Budget"] = summary_by_school_df["Per Student Budget"].map("${:.2f}".format)
summary_by_school_df["% Passing Math"] = summary_by_school_df["% Passing Math"].map("{:.2f}%".format)
summary_by_school_df["% Passing Reading"] = summary_by_school_df["% Passing Reading"].map("{:.2f}%".format)
summary_by_school_df["% Overall Pass Rate"] = summary_by_school_df["% Overall Pass Rate"].map("{:.2f}%".format)

summary_by_school_df[["School Type","Total Students","Total School Budget","Per Student Budget","Avg Reading Score","Avg Math Score", "% Passing Reading", "% Passing Math","% Overall Pass Rate" ]]
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school

```python

math_ninth = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
math_tenth = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
math_eleventh = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
math_twelfth = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

math_by_grade_df = pd.DataFrame({"9th" : math_ninth,
                              "10th" : math_tenth,
                              "11th" : math_eleventh,
                              "12th" : math_twelfth
})
math_by_grade_df["9th"] = math_by_grade_df["9th"].map("{:.2f}".format)
math_by_grade_df["10th"] = math_by_grade_df["10th"].map("{:.2f}".format)
math_by_grade_df["11th"] = math_by_grade_df["11th"].map("{:.2f}".format)
math_by_grade_df["12th"] = math_by_grade_df["12th"].map("{:.2f}".format)

math_by_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

summary_by_school_df["Student_budget"] = summary_by_school_df["Per Student Budget"].str.replace('$',' ')
summary_by_school_df["Student_budget_num"] = pd.to_numeric(summary_by_school_df["Student_budget"])

summary_by_school_df["Spending Ranges"] = pd.cut(summary_by_school_df["Student_budget_num"], spending_bins, labels=group_names)

spending_df = summary_by_school_df.groupby('Spending Ranges')

avg_math_s = spending_df["unformatted Avg Math Score"].mean()
avg_reading_s = spending_df["unformatted Avg Reading Score"].mean()
pass_math_s = spending_df["unformatted % Passing Math"].mean()
pass_reading_s = spending_df["unformatted % Passing Reading"].mean()
overall_pass_s = spending_df["% Overall Pass Rate w/out truncation"].mean()

spending_df = pd.DataFrame({
        "Average Math Score" : avg_math_s,
        "Average reading Score":avg_reading_s,
        "% Passing Math":pass_math_s,
        "% Passing Reading" : pass_reading_s,
        "'% Overall Passing Rate" : overall_pass_s
    
})
spending_df
```
